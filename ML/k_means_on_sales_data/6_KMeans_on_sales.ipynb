```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
try:
    import opendatasets as od
    od.download("https://www.kaggle.com/datasets/kyanyoga/sample-sales-data")
    df = pd.read_csv('sample-sales-data/sales_data_sample.csv', encoding='unicode_escape')
    print("✓ Dataset loaded from Kaggle")
except:
    df = pd.read_csv('Dataset/5-sales_data_sample.csv', encoding='unicode_escape')
    print("✓ Dataset loaded from local directory")
```

```python
to_drop = ['ADDRESSLINE1', 'ADDRESSLINE2', 'STATE', 'POSTALCODE', 'PHONE']
df = df.drop(to_drop, axis=1)
```

```python
#Check for null values
df.isnull().sum()
```

```python
df.dtypes
```

```python
df_numeric = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_numeric)
plt.title("Outlier Detection using Boxplots")
plt.show()

Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))).sum()
print("\nNumber of Outliers per Feature:\n", outliers)

# Cap outliers at Q1 - 1.5*IQR and Q3 + 1.5*IQR
df_capped = df_numeric.copy()
for col in df_numeric.columns:
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]
    df_capped[col] = df_numeric[col].clip(lower=lower_bound, upper=upper_bound)

print("\nOutliers capped successfully.")
print("Capped data shape:", df_capped.shape)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_capped)
print("Data normalized using StandardScaler.")
```

```python
df_normalized = pd.DataFrame(X_scaled, columns=df_capped.columns)

print("\nSample of Normalized Data:")
display(df_normalized.head())

print("\nMean:\n", df_normalized.mean())
print("\nStd:\n", df_normalized.std())
```

```python
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, marker='o')
plt.title("Elbow Method to Determine Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    sil_score = silhouette_score(X_scaled, labels)
    print(f"K = {k} → Silhouette Score = {sil_score:.4f}")
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df_capped['SALES'],
    y=df_capped['MSRP'],
    hue=labels,
    palette='Set2'
)
plt.title("K-Means Clustering Visualization (K = 3)")
plt.xlabel("Sales")
plt.ylabel("MSRP")
plt.legend(title='Cluster')
plt.show()
```
